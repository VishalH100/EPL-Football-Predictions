<h1>Combining football-data and making changes to the data

```python
import constants as c
import pandas as pd
```

```python
# Load the CSV files
df1 = pd.read_csv(c.fb_data_18_19)
df2 = pd.read_csv(c.fb_data_19_20)
df3 = pd.read_csv(c.fb_data_20_21)
df4 = pd.read_csv(c.fb_data_21_22)
df5 = pd.read_csv(c.fb_data_22_23)

# Combine all the dataframes
combined_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

# Rename columns
combined_df.rename(columns={'HomeTeam': 'Home_Team', 'AwayTeam': 'Away_Team'}, inplace=True)

# Define reverse mapping dictionary from df
team_mapping = {
    'Crystal Palace': 'Crystal Palace',
    'Chelsea': 'Chelsea',
    'Man United': 'Manchester Utd',
    'Arsenal': 'Arsenal',
    'Southampton': 'Southampton',
    'Everton': 'Everton',
    'Leicester': 'Leicester City',
    'Aston Villa': 'Aston Villa',
    'Leeds': 'Leeds United',
    'Brentford': 'Brentford',
    'Brighton': 'Brighton',
    'Newcastle': 'Newcastle Utd',
    'West Ham': 'West Ham',
    'Man City': 'Manchester City',
    'Tottenham': 'Tottenham',
    'Liverpool': 'Liverpool',
    'Fulham': 'Fulham',
    'Bournemouth': 'Bournemouth',
    'Wolves': 'Wolves',
    "Nott'm Forest": "Nott'ham Forest",
    'Norwich': 'Norwich City',
    'Burnley': 'Burnley',
    'Watford': 'Watford',
    'Sheffield United': 'Sheffield Utd',
    'West Brom': 'West Brom',
    'Huddersfield': 'Huddersfield',
    'Cardiff': 'Cardiff City'
}

# Map values in combined_df using the mapping dictionary
combined_df['Home_Team'] = combined_df['Home_Team'].map(team_mapping)
combined_df['Away_Team'] = combined_df['Away_Team'].map(team_mapping)


# Save the combined dataframe to a new CSV file
combined_df.to_csv(c.combined_df_filepath, index=False)

combined_df.head(5)
```

<h1>Combining Kaggle and Football-data

```python
import pandas as pd

# Read files into pandas dataframes
file1_df = pd.read_csv(c.combined_df_filepath, encoding ='latin')
file2_df = pd.read_csv(c.EPL_2018_2023, encoding='latin')

# Merge dataframes based on Date, Home_Team, and Away_Team
merged_df = pd.merge(file2_df, file1_df[['Season','Home_Team', 'Away_Team', 'B365H', 'B365D', 'B365A']], 
                    on=['Season','Home_Team', 'Away_Team'], how='left')

# Write merged dataframe back to file2
merged_df.to_csv(c.final_data, index=False)
merged_df
```
